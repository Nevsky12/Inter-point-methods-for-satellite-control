```python
import autograd.numpy as anp
from autograd import grad, jacobian, hessian
from numpy.linalg import solve, norm

class NLPProblem:
    def __init__(self, f, h, g, epsilon=1e-6):
        self.f = f
        self.h = h
        self.g = g
        self.epsilon = epsilon
        self._grad_f = grad(f)
        self._jac_h = jacobian(h)
        self._jac_g = jacobian(g)

    def grad_f(self, x):
        return self._grad_f(x)

    def jac_h(self, x):
        h_val = self.h(x)
        n = len(x)
        m = len(h_val)
        if m == 0:
            return np.zeros((n, 0))
        J_raw = self._jac_h(x)  # shape (m, n)
        return J_raw.T  # shape (n, m)

    def jac_g(self, x):
        g_val = self.g(x)
        n = len(x)
        p = len(g_val)
        if p == 0:
            return np.zeros((n, 0))
        J_raw = self._jac_g(x)  # shape (p, n)
        return J_raw.T  # shape (n, p)

    def hess_lag(self, x, y, w):
        # Hessian of Lagrangian L(x) = f(x) - y^T h(x) - w^T g(x)
        def L(x_):
            return self.f(x_) - np.dot(y, self.h(x_)) - np.dot(w, self.g(x_))

        return hessian(L)(x)


class ArcSearchIP:
    def __init__(self, problem, x0, delta, delta1, delta2):
        self.prob = problem
        self.x = x0
        m, p = len(problem.h(x0)), len(problem.g(x0))
        self.y = np.zeros(m)
        self.w = np.ones(p)
        self.s = np.maximum(problem.g(x0), 0.0001)
        self.z = self.w.copy()
        self.x0 = self.x.copy()
        self.s0 = self.s.copy()
        self.Z0 = self.z.copy()
        self.v0 = np.concatenate([self.x, self.y, self.w, self.s, self.z]).copy()
        self.delta  = delta
        self.delta1 = delta1
        self.delta2 = delta2
        self.hist = []
        print(self.prob.g(x0))
        assert np.all(self.prob.g(x0) > 0)
        assert np.all(self.w > 0)
        assert np.all(self.s > 0)
        assert np.all(self.z > 0)
        assert np.all(self.z == self.w)

    def solve(self, maxiter=100):
        for _ in range(maxiter):
            if not self.step():
                break
        return (self.x, self.y, self.w, self.s, self.z)

    def k_components(self, gradx, Jh, Jg):
        x, y, w, s, z = self.x, self.y, self.w, self.s, self.z
        rx = gradx - Jh @ y - Jg @ w
        rh = self.prob.h(x)
        rg = self.prob.g(x) - s
        rs = w - z
        rz = z * s
        kappa = norm(np.concatenate([rx, rh, rg, rs, rz]))
        return rx, rh, rg, rs, rz, kappa

    def k_alpha(self, alpha, gradx, Jh, Jg, v, v1, v2):
        x_a, y_a, w_a, s_a, z_a = self.split_dv(v - v1 * np.sin(alpha) + v2 * (1 - np.cos(alpha)), len(self.x), len(self.y), len(self.w))
        rx = gradx - Jh @ y_a - Jg @ w_a
        rh = self.prob.h(x_a)
        rg = self.prob.g(x_a) - s_a
        rs = w_a - z_a
        rz = z_a * s_a
        return np.concatenate([rx, rh, rg, rs, rz])
    
    def precompute(self):
        x, y, w = self.x, self.y, self.w
        gradx = self.prob.grad_f(x)
        Jh = self.prob.jac_h(x)
        Jg = self.prob.jac_g(x)
        return gradx, Jh, Jg
    
    def build_rhs1(self, rx, rh, rg, rs, rz, s, z):
        p = len(s)
        mu = (z @ s) / p
        phi = norm(rx)**2 + norm(rh)**2 + norm(rg)**2 + norm(rs)**2 + norm(rz)**2
        sigma = max(0.01, 0.125 * min(1, phi * p / mu))
        # Predictor-corrector RHS (Eq. 15)
        return np.concatenate([rx, rh, rg, rs, rz - sigma * mu * np.ones(p)])
    
    def step(self):
        # Precompute derivatives
        gradx, Jh, Jg = self.precompute()
        # 1) Eq. 6
        rx, rh, rg, rs, rz, kappa = self.k_components(gradx, Jh, Jg)
        if kappa < self.prob.epsilon: return False
        self.hist.append(kappa)
        # 2) Build KKT matrix (Eq. 15)
        A = self.build_kkt(Jh, Jg, gradx)
        # 3) Predictor-corrector RHS (Eq. 23)
        rhs1 = self.build_rhs1(rx, rh, rg, rs, rz, self.s, self.z)
        dv1 = anp.linalg.solve(A, rhs1)
        x1, y1, w1, s1, z1 = self.split_dv(dv1, len(self.x), len(self.y), len(self.w))
        # 4) Corrector RHS (centering)
        rhs2 = self.build_rhs2(self.x, x1, y1, w1, s1, z1)
        dv2 = anp.linalg.solve(A, rhs2)
        # 5) Update along arc (Eq. 19)
        self.arc_update(dv1, dv2, gradx, Jh, Jg)
        return True
    
    def build_kkt(self, Jh, Jg, gradx):
        x, y, w, s, z = self.x, self.y, self.w, self.s, self.z
        n = len(x); m = len(y); p = len(w)
        H = self.prob.hess_lag(x, y, w)
        S = np.diag(s); Z = np.diag(z)
        # KKT block matrix (Eq. 15)
        return np.block([
            [H,               -Jh,             -Jg,             np.zeros((n,p)), np.zeros((n,p))],
            [Jh.T,            np.zeros((m,m)), np.zeros((m,p)), np.zeros((m,p)), np.zeros((m,p))],
            [Jg.T,            np.zeros((p,m)), np.zeros((p,p)), -np.eye(p),      np.zeros((p,p))],
            [np.zeros((p,n)), np.zeros((p,m)), np.eye(p),       np.zeros((p,p)), -np.eye(p)     ],
            [np.zeros((p,n)), np.zeros((p,m)), np.zeros((p,p)), Z,               S              ]
        ])

    def build_rhs2(self, x, x1, y1, w1, s1, z1):
        n, m, p = len(x), len(y1), len(w1)
        # Compute Hessians of each h_j and g_i at x
        Hh = [hessian(lambda v: self.prob.h(v)[j])(x) for j in range(m)]
        Hg = [hessian(lambda v: self.prob.g(v)[i])(x) for i in range(p)]
        # Eq.23 first block: 2*(sum_j y1[j]*Hh[j]*x1 + sum_i w1[i]*Hg[i]*x1)
        term1 = np.zeros(n)
        for j in range(m): term1 += 2 * y1[j] * (Hh[j] @ x1)
        for i in range(p): term1 += 2 * w1[i] * (Hg[i] @ x1)
        # Eq.23 second block (y): [-(x1^T Hh[j] x1)]_j
        term2 = np.array([-(x1 @ Hh[j] @ x1) for j in range(m)])
        # Eq.23 third block (w): [-(x1^T Hg[i] x1)]_i
        term3 = np.array([-(x1 @ Hg[i] @ x1) for i in range(p)])
        # Eq.23 fourth block (s): zeros of length p
        term4 = np.zeros(p)
        # Eq.23 fifth block (z): -2 * z1 * s1
        term5 = -2 * z1 * s1
        return np.concatenate([term1, term2, term3, term4, term5])
    
    def split_dv(self, dv, n, m, p):
        splits = [n, n+m, n+m+p, n+m+2*p]
        return np.split(dv, splits)

    def arc_update(self, dv1, dv2, gradx, Jh, Jg):
        x, y, w, s, z = self.x, self.y, self.w, self.s, self.z
        v = np.concatenate([x, y, w, s, z])
        n = len(x); m = len(y); p = len(w)
        x1, y1, w1, s1, z1 = self.split_dv(dv1, n, m, p)
        x2, y2, w2, s2, z2 = self.split_dv(dv2, n, m, p)
        
        alpha_w = self.alpha_wave(w, w1, w2, s, s1, s2, self.delta)
        alpha_f = self.alpha_flat(x, x1, x2, s, self.delta1, alpha_w)
        alpha_s = self.alpha_strike(v, dv1, dv2, self.delta2, alpha_f, gradx, Jh, Jg)
        alpha_h = self.alpha_hat(z, z1, z2, s, s1, s2, v, dv1, dv2, alpha_s, gradx, Jh, Jg)
        alpha = min(alpha_w, alpha_f, alpha_s, alpha_h)
        
        sin_a = np.sin(alpha)
        cos_a = np.cos(alpha)
        #print("step: ", sin_a)
        self.x = x - x1*sin_a + x2*(1-cos_a)
        self.w = w - w1*sin_a + w2*(1-cos_a)
        self.s = self.prob.g(self.x)
        self.z = np.copy(self.w)
        if m > 0:
            b = (self.prob.grad_f(self.x) - Jg @ self.w)
            self.y = solve(Jh, b)

    def max_step_ith(self, w, w1, w2, delta):
        w = np.asarray(w)
        w1 = np.asarray(w1)
        w2 = np.asarray(w2)
        
        result = np.full_like(w, np.pi / 2, dtype=np.float64)
    
        # Case 1: w1 == 0 and w2 != 0
        mask1 = (w1 == 0) & (w2 != 0)
        dw1 = w - delta * w + w2
        result[mask1] = np.where(dw1[mask1] >= 0, np.pi / 2, np.arccos(dw1[mask1] / w2[mask1]))

        
        # Case 2: w2 == 0 and w1 != 0
        mask2 = (w2 == 0) & (w1 != 0)
        dw2 = w - delta * w
        result[mask2] = np.where(w1[mask2] <= dw2[mask2], np.pi / 2, np.arcsin(dw2[mask2] / w1[mask2]))
        
        # Case 3: w1 > 0 and w2 > 0
        mask3 = (w1 > 0) & (w2 > 0)
        ws3 = np.sqrt(w1**2 + w2**2)
        dw3 = w - delta * w + w2
        result[mask3] = np.where(
            dw3[mask3] >= ws3[mask3],
            np.pi / 2,
            np.arcsin(dw3[mask3] / ws3[mask3]) - np.arcsin(w2[mask3] / ws3[mask3])
        )
    
        # Case 4: w1 > 0 and w2 < 0
        mask4 = (w1 > 0) & (w2 < 0)
        ws4 = ws3  # same as before
        result[mask4] = np.where(
            dw3[mask4] >= ws4[mask4],
            np.pi / 2,
            np.arcsin(dw3[mask4] / ws4[mask4]) + np.arcsin(-w2[mask4] / ws4[mask4])
        )
    
        # Case 5: w1 < 0 and w2 < 0
        mask5 = (w1 < 0) & (w2 < 0)
        ws5 = np.sqrt(w1**2 + w2**2)
        dw5 = w - delta * w + w2
        result[mask5] = np.where(
            dw5[mask5] >= 0,
            np.pi / 2,
            np.pi - np.arcsin(-dw5[mask5] / ws5[mask5]) - np.arcsin(-w2[mask5] / ws5[mask5])
        )
    
        return result

    def alpha_wave(self, w, w1, w2, s, s1, s2, delta):
        alpha_w = self.max_step_ith(w, w1, w2, delta)
        alpha_s = self.max_step_ith(s, s1, s2, delta)
        return np.min(np.minimum(np.minimum(alpha_w, alpha_s), np.pi / 2))

    def find_alpha_flat_hat(self, g_func, xk_func, sk, delta1, alpha_w, tol=1e-4, max_iter=500):
        lo, hi = 0.0001, alpha_w
        for _ in range(max_iter):
            mid = (lo + hi) / 2
            x_mid = xk_func(mid)
            if np.all(g_func(x_mid) >= delta1 * sk):
                lo = mid  # increase alpha_bar
            else:
                hi = mid  # reduce alpha_bar
            if hi - lo < tol:
                break
        return lo

    def alpha_flat(self, x, x1, x2, s, delta1, alpha_w):
        g_func = lambda arg: self.prob.g(arg)
        x_func = lambda alpha: x - x1 * np.sin(alpha) + x2 * (1 - np.cos(alpha))
        return self.find_alpha_flat_hat(g_func, x_func, s, delta1, alpha_w)

    def find_alpha_strike(self, phi_func, v_func, v, delta2, alpha_f_h, tol=1e-4, max_iter=500):
        lo, hi = 0.0001, alpha_f_h
        for _ in range(max_iter):
            mid = (lo + hi) / 2
            v_mid = v_func(mid)
            if np.all(phi_func(v_mid) < phi_func(v) - delta2):
                lo = mid  # increase alpha_bar
            else:
                hi = mid  # reduce alpha_bar
            if hi - lo < tol:
                break
        return lo

    def alpha_strike(self, v, v1, v2, delta2, alpha_f_h, gradx, Jh, Jg):
        phi_func = lambda arg: arg @ arg
        v_func = lambda alpha: self.k_alpha(alpha, gradx, Jh, Jg, v, v1, v2)
        return self.find_alpha_strike(phi_func, v_func, v, delta2, alpha_f_h)

    def find_alpha_hat(self, Z0, s0, v0, phi_func, Zk_func, sk_func, vk_func, alpha_str, tol=1e-4, max_iter=500):
        C = 0.5 * np.dot(Z0, s0) / phi_func(v0)
        
        lo, hi = 0.0001, alpha_str
        for _ in range(max_iter):
            mid = (lo + hi) / 2
            
            Z_alpha = Zk_func(mid)
            s_alpha = sk_func(mid)
            v_alpha = vk_func(mid)
            
            phi_alpha = phi_func(v_alpha)
            m_hat = np.min(Z_alpha @ s_alpha) - C * np.min(phi_alpha)
            
            if m_hat >= 0:
                lo = mid  # increase alpha_hat
            else:
                hi = mid  # reduce alpha_hat
            
            if hi - lo < tol:
                break
        
        return lo

    def alpha_hat(self, Z, Z1, Z2, s, s1, s2, v, v1, v2, alpha_str, gradx, Jh, Jg):
        phi_func = lambda arg: arg @ arg
        Zk_func = lambda alpha: Z - Z1 * np.sin(alpha) + Z2 * (1 - np.cos(alpha))
        sk_func = lambda alpha: s - s1 * np.sin(alpha) + s2 * (1 - np.cos(alpha))
        vk_func = lambda alpha: self.k_alpha(alpha, gradx, Jh, Jg, v, v1, v2)
        return self.find_alpha_hat(self.Z0, self.s0, self.v0, phi_func, Zk_func, sk_func, vk_func, alpha_str)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# HS19 with autograd
def f(x): return (x[0] - 10) ** 3 + (x[1] - 20) ** 3


def h(x): return anp.zeros(0)


def g(x): return anp.array([(x[0] - 5) ** 2 + (x[1] - 5) ** 2 - 100,
                            -((x[1] - 5) ** 2 + (x[0] - 6) ** 2) + 82.81,
                            x[0] - 13,  # x1 >= 13
                            100 - x[0],  # x1 <= 100
                            x[1],  # x2 >= 0
                            100 - x[1]])  # x2 <= 100


x0 = anp.array([15.01, 5.2])
prob = NLPProblem(f, h, g)
solver = ArcSearchIP(prob, x0, 1e-2, 1e-6, 1e-6)
x_opt, y_opt, w_opt, s_opt, z_opt = solver.solve(4000)
print('x*=', x_opt)
print('f(x*)=', f(x_opt))
plt.semilogy(solver.hist, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Residual norm')
plt.title('Convergence HS19 with Autograd')
plt.grid(True)
plt.show()
```
